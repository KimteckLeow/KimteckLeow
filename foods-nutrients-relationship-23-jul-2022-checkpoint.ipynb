```python
# For each item in list, add 10 and print new value
# for x in num_list:
#     x += 10
#     print("The value of the variable 'x' is:", x)

import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import earthpy as et
import json

# Plotting options
sns.set(font_scale=1.3)
sns.set_style("white")

# Set your working directory 
os.chdir(os.path.join(et.io.HOME,
                      'earth-analytics',
                      'food-nutrients'))

daily_food_nutrients_db = json.load(open('src/data/foods.json'))

# Create a list outside of the for loop to store nutrients data

# fnuts = pd.DataFrame(rec['nutrients'], index= range(0,109,1))
# fnuts = pd.DataFrame(rec['nutrients'], index= [0,3])

nutrients = []
for rec in daily_food_nutrients_db:
    fnuts = pd.DataFrame(rec['nutrients'], index=[0])
    fnuts['name'] = rec['name']
# Append the data to the list you made to prevent it from being overwritten    
    nutrients.append(fnuts)

# Combine all the dataframes stored in the nutrients list into one pandas dataframe

pd.set_option('display.max_rows', None)  # or maximum of 1000 rows
nutrients = pd.concat(nutrients, ignore_index=True)
nutrients

#create DataFrame by passing scalar values and passing index
# df = pd.DataFrame({'A': a, 'B': b, 'C': c, 'D': d}, index=[0])
# pd.set_option('display.max_rows', None)  # or maximum of 1000 rows
# fnuts = pd.DataFrame(rec['nutrients'], index= [0,3])
# fnuts
```

```python
#remove duplicates
nutrients = nutrients.drop_duplicates()
nutrients
```

```python
#merging data
nutrient_data = pd.merge(nutrients, info, on='name', how='outer')
nutrient_data
```

```python
# to change the order of DataFrame columns accordingly (name is in the first column)

nutrient_data = nutrient_data[['name','usda_id', 'calories','fat','proteins',
       'carbohydrates', 'serving','Calcium', 'Iron', 'Potassium', 'Magnesium', 'Phosphorus', 'Sodium',
       'Zinc', 'Copper', 'Manganese', 'Selenium', 'Vitamin C', 'Vitamin B12',
       'Vitamin B1', 'Vitamin B2', 'Vitamin B3', 'Vitamin B5', 'Vitamin B6',
       'Vitamin B7', 'Vitamin B9', 'Choline', 'Vitamin A', 'Vitamin E',
       'Vitamin K', 'Linoleic acid', 'Î±-Linolenic acid', 'Tryptophan',
       'Threonine', 'Isoleucine', 'Leucine', 'Lysine', 'Methionine',
       'Phenylalanine', 'Valine', 'Histidine', 'Iodine', 'Molybdenum',
       'Chromium', 'Chloride']]

nutrient_data
```

```python
#drop duplicates in the nutrients DataFrame
nutrient_data.duplicated().sum() # number of duplicates
```

```python
#Select data for a nutrient constitient(protein), rows equal to "Protein" in the column of nutrient
nutrient_data_protein = nutrient_data[nutrient_data["proteins"] > 0 ]
nutrient_data_protein
```

```python
# Sort in descending order for protein contents
nutrient_data_protein_desc = nutrient_data_protein.sort_values(by="proteins", ascending=False)
nutrient_data_protein_desc
```

```python
#display top 25 food id with highest protein contents
pd.set_option('display.max_rows', None)  #to display up to 1000 rows
top25_food_protein = nutrient_data_protein_desc.head(25)
top25_food_protein
```

```python
#Select data for a nutrient constitient(carbohydrates), rows equal to "carbohydrates" in the column of nutrient
nutrient_data_carbohydrates = nutrient_data[nutrient_data["carbohydrates"] > 0 ]
nutrient_data_carbohydrates
```

```python
# Sort in descending order for carbohydrates contents
nutrient_data_carbohydrates_desc = nutrient_data_carbohydrates.sort_values(by="carbohydrates", ascending=False)
nutrient_data_carbohydrates_desc
```

```python
#display top 25 food id with highest carbohydrates contents
pd.set_option('display.max_rows', None)  #to display up to 1000 rows
top25_food_carbohydrates = nutrient_data_carbohydrates_desc.head(25)
top25_food_carbohydrates
```

```python
#Select data for a nutrient constitient(fat), rows equal to "carbohydrates" in the column of nutrient
nutrient_data_fat = nutrient_data[nutrient_data["fat"] > 0 ]
nutrient_data_fat
```

```python
# Sort in descending order for fat contents
nutrient_data_fat_desc = nutrient_data_fat.sort_values(by="fat", ascending=False)
nutrient_data_fat_desc
```

```python
#display top 25 food id with highest fat contents
pd.set_option('display.max_rows', None)  #to display up to 1000 rows
top25_food_fat = nutrient_data_fat_desc.head(25)
top25_food_fat
```

```python
# Import pyplot
import matplotlib.pyplot as plt

      
# Define plot space
fig, ax = plt.subplots(figsize=(30, 15))

# Define x and y axes
# plt.yscale("log")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

ax.bar(top25_food_fat['name'], 
       top25_food_fat['fat'],
       color = 'orange', 
       edgecolor = 'darkblue')

# Set plot title and axes labels
ax.set(title = "Top 25 of product visits in 2015",
       xlabel = "Food name",
       ylabel = "Fat contents")

ax.set(ylim=[0, 1000000])

plt.title('Top 25 of food products with fat contents', fontdict={'fontsize':50})
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.xticks(rotation = 45)

plt.show()
```
